```python
import pandas as pd
```

```python
# 获取数据
order_products = pd.read_csv("D:\\CMY_CODE\PC\\fast-learn-ML\\resource\\instacart\\order_products__prior.csv")
products = pd.read_csv("D:\\CMY_CODE\PC\\fast-learn-ML\\resource\\instacart\\products.csv")
orders = pd.read_csv("D:\\CMY_CODE\PC\\fast-learn-ML\\resource\\instacart\\orders.csv")
aisles = pd.read_csv("D:\\CMY_CODE\PC\\fast-learn-ML\\resource\\instacart\\aisles.csv")
```

```python
# 合并表
table1 = pd.merge(aisles,products,on=["aisle_id","aisle_id"])
```

```python
table2 = pd.merge(table1, order_products, on=["product_id","product_id"])
```

```python
table3 = pd.merge(table2,orders,on=["order_id","order_id"])
```

```python
table3.head()
```

```python
# 找到user_id和aisles之间的关系
table = pd.crosstab(table3["user_id"],table3["aisle"])
```

```python
table
```

```python
data = table[:1000]
```

```python
# PCA降维
from sklearn.decomposition import PCA
```

```python
transfer = PCA(n_components=0.95)
data_new = transfer.fit_transform(data)
data_new.shape
```
